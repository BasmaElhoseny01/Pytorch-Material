## Cross Entropy



```python
import torch
import torch.nn as nn
import numpy as np

loss=nn.CrossEntropyLoss()

Y=torch.tensor([0]) # Gold o/p of the example is Class 0

# nsamples * nclasses = 1*3
Y_pred_good=torch.tensor([[2.0,1.0,0.1]])
Y_pred_bad=torch.tensor([[0.5,2.0,0.3]])

l1=loss(Y_pred_good,Y)
l2=loss(Y_pred_bad,Y)

print(l1.item())
print(l2.item())

# Get the actual Prediction
_,predictions1 = torch.max(Y_pred_good,dim=1)
_,predictions2 = torch.max(Y_pred_bad,dim=1)

print(predictions1)
print(predictions2)

# ##################################################################
# 3 Samples not 1
Y=torch.tensor([2,0,1]) # Gold o/p of the examples


Y_pred_good=torch.tensor([[0.1,1.0,2.1],[2.0,1.0,0.1],[0.1,3.0,0.1]])
Y_pred_bad=torch.tensor([[2.1,1.0,0.1],[0.1,1.0,2.1],[0.1,3.0,0.1]])

l1=loss(Y_pred_good,Y)
l2=loss(Y_pred_bad,Y)


print(l1.item())
print(l2.item())

# Get the actual Prediction
_,predictions1 = torch.max(Y_pred_good,dim=1)
_,predictions2 = torch.max(Y_pred_bad,dim=1)

print(predictions1)
print(predictions2)


```

## MultiClass Problem 


```python
import torch 
import torch.nn as nn

# MultiClass Problem
class NeuralNet2(nn.Module):
    def __init__(self,input_size,hidden_size,num_classes):
        super(NeuralNet2,self).__init__()
        self.linear1=nn.Linear(input_size,hidden_size)
        self.relu=nn.ReLU()
        self.linear2=nn.Linear(hidden_size,num_classes)

    def forward(self,x):
        out=self.linear1(x)
        out=self.relu(out)
        out=self.linear2(x)

        # NO Softmax at the end bec it is in CrossEntropy loss
        return out

model=NeuralNet2(input_size=28*28,hidden_size=5,num_classes=3)
criterion=nn.CrossEntropyLoss() #applies Softmax :D
```

## Binary Classification Problem  


```python
import torch 
import torch.nn as nn

# MultiClass Problem
class NeuralNet1(nn.Module):
    def __init__(self,input_size,hidden_size):
        super(NeuralNet1,self).__init__()
        self.linear1=nn.Linear(input_size,hidden_size)
        self.relu=nn.ReLU()
        self.linear2=nn.Linear(hidden_size,1)

    def forward(self,x):
        out=self.linear1(x)
        out=self.relu(out)
        out=self.linear2(x)
        # Add Sigmoid
        y_pred=torch.sigmoid(out)
        return y_pred

model=NeuralNet1(input_size=28*28,hidden_size=5)
criterion=nn.BCELoss()
```
