```python
import torch
x=torch.tensor(1.0)
y=torch.tensor(2.0)

w=torch.tensor(1.0,requires_grad=True) # Bec this is what we want to optimize :D

# Forward Pass and Compute the loss
y_hat=w*x
loss=(y_hat-y)**2

print(loss)

# Backward Pass
loss.backward()
print(w.grad)

# Next Steps will be to update weights thn next forward ans backward again
```
